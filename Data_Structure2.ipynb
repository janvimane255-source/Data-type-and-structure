Q 17 Describe the time complexity of accessing elements in a dictionary.

- The average-case time complexity of accessing element in dictionary is o(1) (constant time.). This is because dictionarise use hash tables, allowing for direct access to element based in their hash value. In the worst-case scenario 9due to hash collisions), the time complexity can degrade to O(n) (linear time )' but this is rare in practice with well-designed hash functions.
